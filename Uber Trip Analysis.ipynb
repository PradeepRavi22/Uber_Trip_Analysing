```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
uber_read= pd.read_csv("uber-raw-data.csv")
uber_read.head(5)
```

```python
uber_read.tail()
```

```python
uber_read['Date/Time']= pd.to_datetime(uber_read['Date/Time'])
uber_read["Day"] = uber_read["Date/Time"].apply(lambda x: x.day)
uber_read["Hour"] = uber_read["Date/Time"].apply(lambda x: x.hour)
uber_read["Weekday"] = uber_read["Date/Time"].apply(lambda x: x.weekday())
uber_read.head(5)
```

```python
#### Now we can check the density of rides according to days, hours, and weekdays
```

```python
fig,ax = plt.subplots(figsize = (12,6))
plt.hist(uber_read.Day, width= 0.6, bins=30)
plt.title("Density of trips per Day", fontsize=17)
plt.xlabel("day", fontsize=15)
plt.ylabel("Density of rides", fontsize=15)
```

```python
## From the above plot we can notice that the highest number of rides are during working days (Monday to Friday), while the least number of rides are in weekends
```

```python
fig,ax = plt.subplots(figsize = (12,6))
plt.hist(uber_read.Weekday, width= 0.6, range= (0,6.5), bins=7, color= "pink")
plt.title("Density of trips per Weekday", fontsize=17)
plt.xlabel("Weekday", fontsize=15)
plt.ylabel("Density of rides", fontsize=15)
```

```python
fig,ax = plt.subplots(figsize = (12,6))
plt.hist(uber_read.Hour, width= 0.6, bins=24, color= "green")
plt.title("Density of trips per Hour", fontsize=17)
plt.xlabel("Hour", fontsize=15)
plt.ylabel("Density of rides", fontsize=15)
```

```python
#### It represents like the number of rides decrease gradually from 1 AM to 4 PM and then increases starting from 5 AM onward till it reaches 6 PM which is the hour with the highest number of rides.
```

```python
#Visualize the Density of rides per location
fig,ax = plt.subplots(figsize = (12,6))
x= uber_read.Lon
y= uber_read.Lat
plt.scatter(x, y, color="grey")
plt.title("Density of trips per Hour", fontsize=17)
plt.xlabel("Hour", fontsize=15)
plt.ylabel("Density of rides", fontsize=15)
```

```python
# The region with the highest density of rides is near Manhattan and Newburgh. While the region with the lowest density is near New Jersey.
```
